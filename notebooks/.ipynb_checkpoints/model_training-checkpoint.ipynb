```python

# STEP 1: IMPORT LIBRARIES

import pandas as pd
import random
from datetime import datetime, timedelta


# STEP 2: DEFINE CATEGORIES

categories = {
    "Consumer Law": [
        "Seller refused to refund defective product",
        "Received damaged item from online store",
        "Company is not honoring warranty",
        "Wrong product delivered and no replacement",
        "Service provider charged extra money"
    ],
    
    "Cyber Crime": [
        "My bank account was hacked",
        "Received phishing email asking for OTP",
        "Someone misused my social media account",
        "Online fraud transaction happened",
        "Fake website stole my card details"
    ],
    
    "Property Dispute": [
        "Neighbor illegally occupied my land",
        "Builder did not give possession of flat",
        "Property documents are disputed",
        "Tenant is not vacating house",
        "Illegal construction near my property"
    ],
    
    "Domestic Violence": [
        "Husband physically abused me",
        "Facing mental harassment at home",
        "Family member threatening me",
        "Domestic abuse by partner",
        "Unsafe living conditions due to violence"
    ]
}


# STEP 3: TEXT VARIATION FUNCTION

def generate_variation(text):
    variations = [
        text,
        f"I am facing an issue where {text.lower()}",
        f"This is regarding that {text.lower()}",
        f"I want to complain that {text.lower()}",
        f"Please help as {text.lower()}",
        f"{text.lower()} and no action has been taken",
        f"Kindly look into the matter: {text.lower()}",
        f"I am writing to report that {text.lower()}",
        f"This problem occurred because {text.lower()}",
        f"I need help since {text.lower()}"
    ]
    return random.choice(variations)


# STEP 4: DATE GENERATION

def random_date():
    start_date = datetime(2022, 1, 1)
    end_date = datetime(2024, 12, 31)
    delta = end_date - start_date
    random_days = random.randint(0, delta.days)
    return (start_date + timedelta(days=random_days)).strftime("%Y-%m-%d")

# STEP 5: REGION LIST

regions = [
    "Mumbai", "Delhi", "Bangalore", "Chennai",
    "Kolkata", "Pune", "Hyderabad", "Ahmedabad"
]


# STEP 6: GENERATE DATASET

data = []

for category, sentences in categories.items():
    for _ in range(500):  # 500 samples per category
        
        # pick base sentence
        base = random.choice(sentences)
        
        # create variation
        complaint = generate_variation(base)
        
        # generate date and region
        date = random_date()
        region = random.choice(regions)
        
        # append row
        data.append([complaint, category, date, region])


# STEP 7: CREATE DATAFRAME

df = pd.DataFrame(data, columns=["complaint", "category", "date", "region"])


# STEP 8: SAVE DATASET

df.to_csv("../data/synthetic_grievance_dataset.csv", index=False)


# STEP 9: VERIFY OUTPUT

print("Dataset Shape:", df.shape)
print("\nSample Data:")
print(df.head())

print("\nCategory Distribution:")
print(df['category'].value_counts())
```
